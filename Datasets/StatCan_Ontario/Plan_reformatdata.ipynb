# Employment insurance claims received by province and territory, monthly, seasonally adjusted

Retrieved from https://www150.statcan.gc.ca/t1/tbl1/en/cv.action?pid=1410000501


```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./EI_claims.csv', on_bad_lines='skip')
df
```

```python
#Pull necessary columns
df_in = df[['REF_DATE', 'GEO', 'VALUE']]

#Filter to only Ontario
df_in = df_in.loc[df_in['GEO'] == 'Ontario']
df_in = df_in.drop(columns='GEO')
df_in

df_in = df_in.rename(columns={'REF_DATE':'ds', 'VALUE':'statcan_ei_claims'})
df_in
```

```python
df = pd.read_csv('./cpi.csv', on_bad_lines='skip')
df
```

```python
#Define components to look at 
categories= ['All-items', 'Food', 'Shelter']

#Filter to only Ontario
df_in = df.loc[df['Products and product groups'].isin(categories)]
df_in = df_in[['REF_DATE', 'Products and product groups', 'VALUE']]

#Convert each category to individual columns 
df_in = df_in.pivot_table(values='VALUE', columns='Products and product groups', index= df_in.REF_DATE)
df_in = df_in.reset_index()

df_in = df_in.rename(columns={'REF_DATE':'ds', 
                              'All-items':'cpi-allitems',
                              'Food':'cpi-food',
                              'Shelter':'cpi-shelter'
                             })

#Convert to datetime
df_in['ds'] = pd.to_datetime(df_in['ds'])
df_in = df_in.set_index('ds')
df_in
```
