```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
```

```python
df = pd.read_csv('GPT4_Test1outputs.csv')
df

#Using test period 1 where the start date is 2023-03-08
# Start date
start_date = datetime(2023, 3, 8)

# Create a list of 26 datetime objects at 1-week intervals
date_list = [start_date + timedelta(weeks=i) for i in range(26)]

    
df['timestamp'] = date_list
df.set_index('timestamp')

df = df.rename(columns={'Unnamed: 1': 'all_forecasts'})
df
```

```python
#Create table of 1 week forecasts
#split the data into individual lists
data_string = df['all_forecasts'].values[0]
data_list = [list(map(float, sublist.strip('][').split(', '))) for sublist in data_string.split('; ')]

# Access the elements programmatically
print("1-Week Forecast:")
for value in data_list[0]:
    print(value)

print("\n4-Week Forecast:")
for value in data_list[1]:
    print(value)

print("\n12-Week Forecast:")
for value in data_list[2]:
    print(value)
```
