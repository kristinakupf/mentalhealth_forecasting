```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from functools import reduce 
import codecs
import statsmodels.api as sm
import datetime
```

## Load the Statistics Canada Dataset 

```python
dataset = 'StatCan'
DATA_FOLDER = "../raw_data/Datasets/StatCan_Ontario/"

print('Loading and pre-processing {} data for use in forecasting model'.format(dataset))

def filter_Ontario(df):
#     categories= df['Products and product groups'].unique()

    #Filter to only Ontario
#     df_in = df.loc[df['Products and product groups'].isin(categories)]
    df_in = df[['REF_DATE', 'Products and product groups', 'VALUE']]

    #Convert each category to individual columns 
    df_in = df_in.pivot_table(values='VALUE', columns='Products and product groups', index= df_in.REF_DATE)
    df_in = df_in.reset_index()

    df_in = df_in.rename(columns={'REF_DATE':'ds', 
                                 })

    #Convert to datetime
    df_in['ds'] = pd.to_datetime(df_in['ds'])
    df_in = df_in.set_index('ds')
    return df_in

df = pd.read_csv('{}/cpi.csv'.format(DATA_FOLDER), on_bad_lines='skip')

output_df = filter_Ontario(df)


output_df = output_df.rename(columns={old_name: 'statcan_' + old_name for old_name in output_df.columns})
```

```python
output_df.to_csv('../processed_data_inputs/StatCan.csv')
```
