```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from functools import reduce 
import codecs
import statsmodels.api as sm
import datetime
```

## Load the Connex-Ontario Dataset 

```python
def prep_data(data_path, encoding, delimiter):
    
    #Load main 
    main_df = pd.read_csv(data_path+'Main Contacts_en-ca.csv',encoding=encoding, sep=delimiter)
    main_df['DATE'] = [datetime.datetime.strptime(str(d), '%Y%m%d') for d in main_df['CONTACT_DAY']]

    #Target groups
    target_groups_df = pd.read_csv(data_path+'Target Groups_en-ca.csv',encoding=encoding, sep=delimiter)
    target_groups_df = target_groups_df.rename(columns={'TARGET_POPULATION':'TARGET_GROUP'})
    
    #Diagnosis groups
    diagnosis_df = pd.read_csv(data_path+'Diagnosis_en-ca.csv',encoding=encoding, sep=delimiter)

    #Ethnicicty
    ethnicity_df = pd.read_csv(data_path+'Ethnicity_en-ca.csv',encoding=encoding, sep=delimiter)
    ethnicity_df = ethnicity_df.rename(columns={'TARGET_POPULATION':'ETHNICITY'})

    #Languages
    language_df = pd.read_csv(data_path+'Languages_en-ca.csv',encoding=encoding, sep=delimiter)

    #Special needs provisions
    special_needs_df = pd.read_csv(data_path+'Special Provisions_en-ca.csv',encoding=encoding, sep=delimiter)

    #Referrals groups
    referrals_df = pd.read_csv(data_path+'Referrals_en-ca.csv',encoding=encoding, sep=delimiter)
    referrals_df = referrals_df.rename(columns={'FUNCTIONAL_CENTRE':'REFERRAL_CENTRE'})
    referrals_df = referrals_df.rename(columns={'REFERRALS':'NUM_REFERRALS'})
    # referrals_df

    #Supplementary Resources groups
    supplementary_resources_df = pd.read_csv(data_path+'Supplementary Resources_en-ca.csv',encoding=encoding, sep=delimiter)
    supplementary_resources_df = supplementary_resources_df.rename(columns={'RESOURCE_TYPE':'SUPP_RESOURCE_TYPE'})
    supplementary_resources_df = supplementary_resources_df.rename(columns={'REFERRALS':'NUM_SUPP_RESOURCES'})
        
    #Combine all dataframes to single dataframe with more columns
    dfs = [main_df, target_groups_df, diagnosis_df, ethnicity_df, language_df, special_needs_df, referrals_df, supplementary_resources_df]
    df_out = reduce(lambda left, right: pd.merge(left, right, how='left', on='CLIENT_PROFILE_ID'), dfs)
    
    return df_out

dataset = 'CONNEX'
DATA_FOLDER = "../raw_data/Datasets/CONNEX"

print('Loading and pre-processing {} data for use in forecasting model'.format(dataset))


#Get main historical dataframe of data from 2015- 2020
print('Getting {} data from 2015 - 2020'.format(dataset))
main_df = prep_data("{}/CONNEXMAIN_2015_2020/".format(DATA_FOLDER), 'latin1', ',')
main_df

#Add the additional datasets for each month 
dfs = [main_df]

#Combine with modern data 
for data_month in os.listdir(DATA_FOLDER):
    if 'CONNEX_' in data_month:
        print('Getting {} data from {}'.format(dataset, data_month))
        data_path = DATA_FOLDER+data_month+'/'
        
        df_concat = prep_data(data_path,'utf-16', '\t')
        dfs.append(df_concat)
        
result_df = pd.concat(dfs)

#Check that date range looks appropriate 
print('Earliest date in dataset is {}'.format(result_df['DATE'].min()))
print('Earliest date in dataset is {}'.format(result_df['DATE'].max()))
```

```python
#For all populations
def consolidate_data(df, category):
    #Get the total counts for days for the different contact_methods
    date_df = pd.pivot_table(df, values=[category], index=['DATE'], aggfunc=np.sum)
    return date_df
    

categories = ['IS_MENTAL_HEALTH', 'IS_SUBSTANCE_ABUSE', 'IS_PROBLEM_GAMBLING', 'IS_OTHER']

for category in categories:
    if category == categories[0]: #if first time through
        output_df = consolidate_data(result_df, category)
    else:
        output_df[category] = consolidate_data(result_df, category)[category]
        
output_df.plot()
```

### Identify groups that we will include as separate time-series 
Based on this analysis we will use diagnostic group, target group, and ethnicity

```python
#Do descriptive statistics for different groups
# groups = ['TARGET_GROUP', 'DIAGNOSTIC_GROUP', 'ETHNICITY', 'DIAGNOSIS', 'LANGUAGE']
groups = ['TARGET_GROUP', 'DIAGNOSTIC_GROUP', 'ETHNICITY',]

result_df[groups].count()

top_all = []
top_k = 10

for group in groups:
    percent = (result_df[group].count()/len(result_df))*100
    print('{}% of data has {} information'.format(percent, group))
    top = {'group':group, 
             'top_k':result_df[group].value_counts().sort_values(ascending=False).head(top_k)
            }

    top_all.append(top)
```

```python
# For each group create a time-series with the same categories 

counter = 0 
for group in groups:  ## Different metrics to use like ['TARGET_GROUP', 'DIAGNOSTIC_GROUP', 'ETHNICITY',]
    top = top_all[counter]['top_k'].index.values
    print(top)
    counter+=1
    for value in top: ### For each individual group 
        limited_df = result_df.loc[result_df[group].isin([value])]
        
        #Get counts for categories used above
        for category in categories: ### ['IS_MENTAL_HEALTH', 'IS_SUBSTANCE_ABUSE', 'IS_PROBLEM_GAMBLING']
            output_df[category+'_'+ value] = consolidate_data(limited_df, category)[category]

        
output_df       
    
```

```python
# Do some quality control to make sure signals have some data 
output_df = output_df.fillna(0)

output_df.to_csv('../processed_data_inputs/Connex.csv')
```
